```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.huggingface import HuggingFaceLLM

#  simpleDirectoryReader is used to read documents and vectorStoreIndex is used fo indexing the documents
#  into vectors and settings is used to specify that the system should use the specific llm modal and embedding modal
#HuggingFaceEmbedding is used to specify the embedding model  and HuggingFacceLLM is used to specify the LLM modal.
```

```python
#In this cell we are using the file operator in which we will be opening the document that is in write
# formate and in f.write we will be writing the content.

with open("demo.txt", "w") as f:
  f.write(
      """
      A galaxy is a system of stars, stellar remnants, interstellar gas, dust, and dark matter bound together by gravity.
      The word is derived from the Greek galaxias, literally 'milky', a reference to the Milky Way galaxy that contains the Solar System.
      Galaxies, averaging an estimated 100 million stars, range in size from dwarfs with less than a thousand stars,
      to the largest galaxies known – supergiants with one hundred trillion stars, each orbiting its galaxy's centre of mass.
      Most of the mass in a typical galaxy is in the form of dark matter,
      with only a few per cent of that mass visible in the form of stars and nebulae.
      Supermassive black holes are a common feature at the centres of galaxies.
      """
  )
# opens our docs and analyses the info in it by noting it down
```

```python
document=SimpleDirectoryReader(".").load_data() # reads the input doc
```

```python
embed_model=HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
#here we will assign the embeddding model.
```

```python
llm=HuggingFaceLLM(
    model_name="gpt2", #meta-llama/Llama-2-7b-chat-hf
    tokenizer_name="gpt2",
)
```

```python
Settings.embed_model=embed_model         # sets the process/mechanism globally
Settings.llm=llm
```

```python
index=VectorStoreIndex.from_documents(document)  # complete indexing process is done in this step
```

```python
qa=index.as_query_engine()
#updates or changes index as a query engine (that Let's us to ask question and generate answer)
```

```python
import re # to clean the text, removing special chars and compressing the output (regular expression)

def clean_output(response):
  text=str(response).strip()
  first_line=re.split(r"\n", text)[0].strip()
  return first_line

print("\n Answer::", clean_output(qa.query("What is a galaxy?")))

#.quer(...) is a method of the query engine that:
#take your input question.
#Converts it into an embedding (via embed_model)
#Searches the index for similar document chunks.
#Sends thee most relevant context + your  question to the LLM(via LLM).
#Return a response generated by the LLM.
```
